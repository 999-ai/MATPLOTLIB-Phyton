```python
import matplotlib.pyplot as plt

days=["Monday","Thursday","Wednesday","Tuesday","Friday","Saturday","Sunday"]
temperatures=[22,24,19,23,25,20,18]

plt.plot(days,temperatures,color="r")# izgi grafii oluturma
# Balk ve etiketler ekleme
plt.xlabel("days")
plt.ylabel("temperatures (C)")
plt.title("weekly temperature graphic ")
# Izgara izgileri ekleme
plt.grid(True)
# Grafii gsterme
plt.show()
```

```python
import matplotlib.pyplot as plt

days=["Monday","Thursday","Wednesday","Tuesday","Friday","Saturday","Sunday"]
temperatures1=[22,24,19,23,25,20,18]
temperatures2=[25,22,27,26,28,23,21]

plt.plot(days,temperatures1,label="Izmir",color="b",marker="o",linestyle="--")
plt.plot(days,temperatures2,label="Istanbul",color="r",marker="s",linestyle="-")


plt.title("Temperatures in Izmir and Istanbul")
plt.xlabel("days")
plt.ylabel("temperatures(C)")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

categories=["Electronics","Clothing","Food","Furniture","Books"]
sales=[3000,1500,2000,3500,25600]

plt.bar(categories,sales,color="skyblue")
plt.xlabel("categories")
plt.ylabel("sales")
plt.title("Monthly sales in terms of categories(TL)")


for i ,sale in enumerate(sales):
    plt.text(i,sale+100,str(sale+100),ha="center")
```

```python
import matplotlib.pyplot as plt
categories=["Marketing","Researh and Development","Personnel","Office Expenditure","Other"]
budget=[1500,2000,3000,5000,10000]


plt.pie(budget,labels=categories,autopct="%1.1f%%",colors=["gold","lightcoral","lightskyblue","lightgreen","lightpink"])# Pasta grafii oluturma


plt.title("Budget distribution  according to categories")

plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

categories=["A","B","C","D","E"]
values=np.random.randint(10,100,size=len(categories))



plt.bar(categories,values,color="blue")


plt.title("Values interm of categories")
plt.xlabel("Cateories")
plt.ylabel("values")


for i,value in enumerate(values):
 plt.text(i,value+1,str(value+1),fontsize=10,color="red",ha="right")

plt.show()
```
